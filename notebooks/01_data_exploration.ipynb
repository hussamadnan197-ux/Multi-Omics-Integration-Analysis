```python
# 1. إنشاء بيانات تجريبية تحاكي مرضى سرطان الدم AML
np.random.seed(42)
n_patients = 100

data_sim = {
    'Patient_ID': [f'AML_{i:03d}' for i in range(1, n_patients + 1)],
    'Age': np.random.randint(18, 75, size=n_patients),
    'Gender': np.random.choice(['Male', 'Female'], size=n_patients),
    'Survival_Status': np.random.choice(['Alive', 'Dead'], size=n_patients, p=[0.4, 0.6]),
    'Survival_Months': np.random.randint(1, 60, size=n_patients),
    'FLT3_Mutation': np.random.choice(['Positive', 'Negative'], size=n_patients, p=[0.3, 0.7])
}

# 2. تحويل البيانات إلى DataFrame (جدول pandas)
df = pd.DataFrame(data_sim)

# 3. عرض الأسطر الخمسة الأولى من البيانات لاستكشافها
df.head()
```

```python
df.describe()
```

```python
# حساب عدد المرضى المصابين بالطفرة ونسبتهم
print("--- توزيع طفرة FLT3 بين المرضى ---")
print(df['FLT3_Mutation'].value_counts())

print("\n--- توزيع حالة البقاء على قيد الحياة ---")
print(df['Survival_Status'].value_counts())
```

```python
# 1. ضبط حجم الرسم البياني
plt.figure(figsize=(8, 5))

# 2. رسم مخطط بياني عمودي يقارن حالة البقاء لكل مجموعة (طافرة وغير طافرة)
sns.countplot(x='FLT3_Mutation', hue='Survival_Status', data=df, palette='Set2')

# 3. إضافة العناوين والتسميات التوضيحية للمشروع
plt.title('Impact of FLT3 Mutation on Patient Survival Status in AML', fontsize=14, fontweight='bold')
plt.xlabel('FLT3 Mutation Status', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)
plt.legend(title='Survival Status')

# 4. عرض الرسم
plt.show()
plt.savefig('results/FLT3_Mutation_Survival_Status.png', dpi=300, bbox_inches='tight')
```

```python
# Grouping data to calculate the mean survival months for each mutation group
survival_analysis = df.groupby('FLT3_Mutation')['Survival_Months'].mean().reset_index()

print("--- Average Survival Months by FLT3 Mutation Status ---")
print(survival_analysis)

# Plotting the survival months comparison
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
sns.barplot(x='FLT3_Mutation', y='Survival_Months', data=survival_analysis, palette='Blues_r')

plt.title('Average Survival Months based on FLT3 Mutation Status', fontsize=12, fontweight='bold')
plt.xlabel('FLT3 Mutation Status', fontsize=10)
plt.ylabel('Mean Survival Months', fontsize=10)

# Saving the figure to the results folder
plt.savefig('../results/survival_months_by_mutation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Simulate Gene Expression Matrix (RNA-Seq log2 counts for key AML genes)
genes = ['FLT3', 'NPM1', 'DNMT3A', 'IDH1', 'RUNX1']
expression_data = np.random.normal(loc=5.0, scale=1.5, size=(len(genes), n_patients))
df_expr = pd.DataFrame(expression_data, index=genes, columns=df['Patient_ID'])

print("--- Gene Expression Matrix (First 5 Patients) ---")
print(df_expr.iloc[:, :5])

# 2. Simulate DNA Methylation Matrix (Beta-values bounded between 0 and 1)
cpg_sites = ['cg00012345', 'cg00056789', 'cg00098765']
methylation_data = np.random.beta(a=2, b=5, size=(len(cpg_sites), n_patients))
df_meth = pd.DataFrame(methylation_data, index=cpg_sites, columns=df['Patient_ID'])

print("\n--- DNA Methylation Matrix (Beta-values for First 5 Patients) ---")
print(df_meth.iloc[:, :5])

# 3. Save matrices to data folder for future integration steps
df_expr.to_csv('../data/expression_matrix.csv')
df_meth.to_csv('../data/methylation_matrix.csv')
print("\n✅ Both molecular matrices successfully saved to '../data/' directory!")
```

```python
import scipy.stats as stats

# 1. Extract FLT3 expression and a specific CpG site methylation values
flt3_expr = df_expr.loc['FLT3']
cpg_meth = df_meth.loc['cg00012345']

# 2. Calculate Pearson correlation coefficient
corr_coef, p_value = stats.pearsonr(flt3_expr, cpg_meth)
print("--- Multi-Omics Correlation Analysis ---")
print(f"Pearson Correlation Coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4e}")

# 3. Plot the integration scatter plot
plt.figure(figsize=(7, 5))
sns.regplot(x=flt3_expr, y=cpg_meth, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})

plt.title(f'Integration: FLT3 Expression vs cg00012345 Methylation\n(r = {corr_coef:.2f}, p = {p_value:.2e})', fontsize=12, fontweight='bold')
plt.xlabel('FLT3 Gene Expression (log2 counts)', fontsize=10)
plt.ylabel('DNA Methylation Level (Beta-value)', fontsize=10)

# Save the integration plot
plt.savefig('../results/omics_integration_correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```
